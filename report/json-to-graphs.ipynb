```python
import json
import matplotlib.pyplot as plt
import numpy as np


base = '../runs/json/'
```

```python


def plotgraphs(path, ylab):
    with open(path,'r') as file:
        data = json.load(file)

    data = np.array(data)
    # print(data)
    # print(data[:,1])

    # for _, x, y in data:
    plt.plot(data[:,1],data[:,2],'-',c='b')
    plt.xlabel('Epoch')
    plt.ylabel(ylab)
    plt.show()

plotgraphs(base+'ce.json', 'Cross-Entropy Loss')
```
